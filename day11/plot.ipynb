```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tit = sns.load_dataset('titanic')
tit
```

```python
tit.info()
```

```python
tit.isnull().sum()
```

```python
tit.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived', hue = 'sex', data = tit)
```

```python
#scatter Plot

plt.scatter(data=tit, x='age', y = 'fare', c = 'pclass')
```

```python
sns.scatterplot(data=tit, x='age', y = 'fare', hue = 'pclass')
```

```python
sns.barplot(data=tit, x='sex', y = 'fare', hue = 'pclass')
```

```python
tit.columns
```

```python
sns.pairplot(tit)
```

```python
tit.hist(bins = 10, figsize=(9,7), grid=False)
```

```python
sns.boxplot(data=tit, x='sex', y = 'fare', hue = 'pclass',)
```

```python
#how to change x and y axes title
sns.boxenplot(data=tit, x='sex', y = 'fare', hue = 'pclass')
plt.xlabel("Sex")
plt.ylabel("Fare")
```

```python
tit
```

```python
 g= sns.FacetGrid(tit, hue='survived', col='pclass', margin_titles=True) 
#        palette=(1:'seagreen', 0:'gray'))

g= g.map(plt.scatter, 'fare', 'age', edgecolor= 'w').add_legend();

 
```

```python
s=sns.FacetGrid(tit, hue='survived', col='pclass', margin_titles=True,
                    palette = (1:'seagreen', 0:'gray'))
```

```python
sns.stripplot(data=tit, x='sex', y = 'fare', hue = 'class')
```

```python
#Heatmap
corr = tit.corr()
plt.Figure(figsize=(10,10))
sns.heatmap(corr)
plt.title('Heatmap')
```

```python
sns.violinplot(data=tit, x='sex', y = 'fare', hue = 'class')
```
