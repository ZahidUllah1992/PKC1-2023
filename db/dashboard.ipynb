```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe, skipping the first 6 rows and the last row
df = pd.read_csv("ub.csv", skiprows=6, skipfooter=1, engine="python")

# Get a list of all the columns in the dataframe, excluding the "Date" and "GCell Group" columns
columns = [col for col in df.columns if col not in ["Date", "GCell Group"]]

# Create a figure and axis for each column
fig, axs = plt.subplots(len(columns), figsize=(10, 10))

# Loop through each column
for i, column in enumerate(columns):
    # Filter the dataframe to only include the current column
    df_column = df[["Date", "GCell Group", column]]

    # Get a list of all the unique enodeb names
    GCell_Group_names = df["GCell Group"].unique()
    
    # Loop through each enodeb name
    for GCell_Group_name in GCell_Group_names:
        # Filter the data to only include the current enodeb
        df_enodeb = df_column[df_column["GCell Group"] == GCell_Group_name]
        
        # Plot the data on the current axis
        axs[i].plot(df_enodeb["Date"], df_enodeb[column], label=GCell_Group_name)
    
    # Add the x and y labels and title for the current axis
    axs[i].set_xlabel("Date")
    axs[i].set_ylabel("Value")
    axs[i].set_title("Line chart for {}".format(column))
    axs[i].legend()

# Show the figure
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe, skipping the first 6 rows and the last row
df = pd.read_csv("ub.csv", skiprows=6, skipfooter=1, engine="python")

# Get a list of all the unique enodeb names
GCell_Group_names = df["GCell Group"].unique().tolist()

# Create a figure and axis for each enodeb name
fig, axs = plt.subplots(len(df.columns) - 2, sharex=True, figsize=(10,10))

# Loop through each column after the "Date" and "GCell Group" columns
for i, column in enumerate(df.columns[2:]):
    # Loop through each enodeb name
    for j, GCell_Group_name in enumerate(GCell_Group_names):
        # Filter the dataframe to only include the current enodeb
        df_enodeb = df[df["GCell Group"] == GCell_Group_name]

        # Plot the data on the current axis
        axs[i].plot(df_enodeb["Date"], df_enodeb[column], label=GCell_Group_name)
        axs[i].legend()
        axs[i].set_ylabel("Value")
        axs[i].set_title("Line chart for {}".format(column))

# Show the figure
plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.set_page_config(page_title="Charts", page_icon=":chart_with_upwards_trend:", layout="wide")

# Load the data into a pandas dataframe, skipping the first 6 rows and the last row
df = pd.read_csv("ub.csv", skiprows=6, skipfooter=1, engine="python")

# Get a list of all the unique enodeb names
GCell_Group_names = df["GCell Group"].unique().tolist()

# Add a multi-select checkbox for the enodeb names in the sidebar
selected_GCell_Group_names = st.sidebar.multiselect("Select GCell Groups", GCell_Group_names)

# Create a figure and axis for each enodeb name
cols = 2
rows = (len(df.columns) - 2) // cols + (len(df.columns) - 2) % cols
fig, axs = plt.subplots(rows, cols, figsize=(15, 50))
axs = axs.flatten()

# Loop through each column after the "Date" and "GCell Group" columns
for i, column in enumerate(df.columns[2:]):
    # Plot the data on the current axis
    for j, GCell_Group_name in enumerate(selected_GCell_Group_names):
        # Filter the dataframe to only include the selected enodebs
        df_enodeb = df[df["GCell Group"] == GCell_Group_name]

        axs[i].plot(df_enodeb["Date"], df_enodeb[column], label=GCell_Group_name)
        axs[i].set_ylabel("Value")
        axs[i].set_xlabel("Date")
        axs[i].set_xticklabels(df_enodeb["Date"], rotation=90)
        axs[i].set_title(column)



fig.legend(GCell_Group_names, loc='upper center', ncol=len(GCell_Group_names), bbox_to_anchor=(0.5, 1.15))
fig.tight_layout()

# Show the figure
st.pyplot()
```

```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("ub.csv", skiprows=6, skipfooter=1, engine="python")

# plot line chart for CEFR
fig = px.line(df, x='Date', y='CEFR', color="GCell Group",
              hover_data={"Date": "|%B %d, %Y"},
              title='CEFR')
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")
fig.show()

# plot line chart for KPI after Integrity
kpi_cols = df.columns[df.columns.get_loc("Integrity")+1:]
for col in kpi_cols:
    fig = px.line(df, x='Date', y=col, color=df.columns[1],
                  hover_data={"Date": "|%B %d, %Y"},
                  title=col)
    fig.update_xaxes(
        dtick="M1",
        tickformat="%b\n%Y")
    fig.show()
```
